```python
# plotting pairplot using the data
sns.pairplot(data)
plt.show()
```

```python
#plotting heatmap using the data
sns.heatmap(data.corr(numeric_only=True),annot=True)
plt.show()
```

```python
#plotting boxplot classification against SL
sns.boxplot(x='Classification',y='SL',data=data)
```

```python
#plotting boxplot classification against SW
sns.boxplot(x='Classification',y='SW',data=data)
```

```python
#plotting boxplot classification against PL
sns.boxplot(x='Classification',y='PL',data=data)
```

```python
#plotting boxplot classification against PW
sns.boxplot(x='Classification',y='PW',data=data)
```

```python
#grouping the data w.r.t mean
data.groupby('Classification')[['SL','SW','PL','PW']].mean(numeric_only=True)
```

```python
#plotting the grouped data w.r.t mean
data.groupby('Classification')[['SL','SW','PL','PW']].mean(numeric_only=True).plot.bar()
plt.show()
```
