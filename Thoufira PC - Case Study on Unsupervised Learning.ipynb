```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

scaled_data = pd.DataFrame(scaled_data,columns = data.columns)
```

```python
from sklearn.decomposition import PCA
pca=PCA().fit(scaled_data)
plt.rcParams["figure.figsize"]=(12,6)
fig,ax=plt.subplots()
xi=np.arange(1,14,step=1)
y=np.cumsum(pca.explained_variance_ratio_)
plt.ylim(0.0,1.1)
plt.plot(xi,y,marker='o',linestyle='--',color='b')
plt.xlabel('Number of Components')
plt.xticks(np.arange(0,14,step=1))
plt.ylabel('Cumulative variance(%)')
plt.title('The number of components needed to explain variance')
plt.axhline(y=0.97,color='r',linestyle='-')
plt.text(0.5,0.85,'97% cut-off threshold',color='red',fontsize=16)
ax.grid(axis='x')
plt.show()
```

```python
# Perform PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.97)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
# Cluster Models
```

```python
# 1. K-Means Clustering
```

```python
from sklearn.cluster import KMeans

# Finding the Optimum Number of Clusters
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init= 'k-means++', random_state=42)
    kmeans.fit(x_pca)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the elbow curve
plt.plot(range(1,11), wcss)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Curve for K-means")
plt.show()
```

### From the Elbow Curve , the optimum number of clusters = 3

```python
# K-means clustering model
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans_pred=kmeans.fit_predict(x_pca)
```

```python
# Evaluation Metrics
```

```python
from sklearn.metrics import silhouette_score

kmeans_silhouette = silhouette_score(x_pca,kmeans_pred)
print(kmeans_silhouette)
```

```python
# 2. Heirarchical Clustering (Agglomerative)
```

```python
import scipy.cluster.hierarchy as sch 

# Finding the Optimum Number of Clusters
dendrogram= sch.dendrogram(sch.linkage(x_pca,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Number Of Clusters')
plt.ylabel('Euclidean Distances')
plt.show()
```

### From the Dendrogram , the optimum number of clusters = 3

```python
# Agglomerative Clustering Models
from sklearn.cluster import AgglomerativeClustering

agg=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
agg_pred = agg.fit_predict(x_pca)
```

```python
# Evaluation Metrics
```

```python
from sklearn.metrics import silhouette_score

agg_silhouette = silhouette_score(x_pca,agg_pred)
print(agg_silhouette)
```

```python
# 3. DBSCAN Clustering
```

```python
# Finding Optimal Value of Epsilon
from sklearn.neighbors import NearestNeighbors

nearest_neighbors = NearestNeighbors(n_neighbors=23)
neighbors = nearest_neighbors.fit(scaled_data)

distances, indices = neighbors.kneighbors(scaled_data)
k_distances = np.sort(distances[:,22], axis=0)
```

```python
# PLot the nearest neighbor
fig = plt.figure(figsize=(15, 8))
plt.plot(k_distances)
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()
```

```python
# Finding the knee location
from kneed import KneeLocator

knee = KneeLocator(range(1,len(k_distances)+1), k_distances, S = 1.0, curve='convex', direction='increasing')
```

```python
# Plot the knee locator
fig = plt.figure(figsize=(15, 8))
knee.plot_knee()
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()
```

```python
# print the knee locator
print(k_distances[knee.knee])
```

```python
# DBSCAN Clustering Models
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps = 3.887, min_samples = 22)
dbscan_labels = dbscan.fit_predict(scaled_data)
```

```python
# Get the number of clusters (excluding noise)
n_clusters = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
print("Number of clusters:", n_clusters)

# Identify and count the number of noise points
n_noise_points = list(dbscan_labels).count(-1)
print("Number of noise points:", n_noise_points)
```

```python
# Evaluation Metrics
```

```python
from sklearn.metrics import silhouette_score

dbscan_silhouette = silhouette_score(scaled_data, dbscan_labels)
print(dbscan_silhouette)
```

```python
print('K-Means Silhouette Score : ' ,kmeans_silhouette)
print('Agglomerative Heirarchical Silhouette Score : ' ,agg_silhouette)
print('DBSCAN Silhouette Score : ' ,dbscan_silhouette)
```
