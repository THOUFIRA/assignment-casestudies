# 2.

```python
parents=df['parental level of education'].value_counts()
parents
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='parental level of education',data=df)
plt.show()
```

```python
df.groupby(by='parental level of education').mean(numeric_only=True)
```

```python
df.groupby(by=['parental level of education']).mean(numeric_only=True).plot.bar()
plt.show()
```

```python
#Students score tend to increase directly as per their parent's education. Parents having master's degree have a significant impact on students to score more than those parents who only have a high school degree.
```

# 3.

```python
df.groupby('gender')['math score','reading score','writing score'].mean()
```

```python
df.groupby('gender')['math score','reading score','writing score'].mean().plot.bar()
plt.show()
```

```python
df.groupby(['gender','parental level of education'])[['math score','reading score','writing score']].mean().plot.bar()
plt.show()
```

```python
#students with the same parental level of education,female students have higher average reading and writing scores than male.where male students have higher average math score
```

```python
df.groupby('test preparation course')['math score','reading score','writing score'].mean()
```

```python
df.groupby('test preparation course')['math score','reading score','writing score'].mean().plot.bar()
plt.show()
```

```python
df.groupby(['test preparation course','parental level of education'])[['math score','reading score','writing score']].mean().plot.bar()
plt.show()
```

```python
# students who have completed the test preparation course and with a bachelor's degree as the parental level of education have higher average math scores
# students who have completed the test preparation course and with a master's degree as the parental level of education have higher average reading scores and writing scores
```
